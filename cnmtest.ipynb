```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pytesseract
plt.style.use('dark_background')
```

```python
img_ori_1 = cv2.imread('7.png')
img_ori_2 = cv2.imread('8.png')
img_ori_3 = cv2.imread('9.png')
img_ori_4 = cv2.imread('10.png')

gray1 = cv2.cvtColor(img_ori_1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img_ori_2, cv2.COLOR_BGR2GRAY)
gray3 = cv2.cvtColor(img_ori_3, cv2.COLOR_BGR2GRAY)
gray4 = cv2.cvtColor(img_ori_4, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(12, 10))
plt.imshow(img_ori_1,cmap='gray')
```

```python
kernel5 = np.ones((5,5),np.uint8)
img_morphology5_blur = cv2.GaussianBlur(gray1, ksize=(5, 5), sigmaX=0)
img_morphology5_open= cv2.morphologyEx(img_morphology5_blur, cv2.MORPH_OPEN, kernel5)
img_morphology5_close = cv2.morphologyEx(img_morphology5_open, cv2.MORPH_CLOSE, kernel5)
img_morphology5_thresh_1 = cv2.adaptiveThreshold(
    img_morphology5_close,
    maxValue=255.0,
    adaptiveMethod=cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    thresholdType=cv2.THRESH_BINARY_INV,
    blockSize=19,
    C=9)
img_morphology5_blur = cv2.GaussianBlur(gray2, ksize=(5, 5), sigmaX=0)
img_morphology5_open= cv2.morphologyEx(img_morphology5_blur, cv2.MORPH_OPEN, kernel5)
img_morphology5_close = cv2.morphologyEx(img_morphology5_open, cv2.MORPH_CLOSE, kernel5)
img_morphology5_thresh_2 = cv2.adaptiveThreshold(
    img_morphology5_close,
    maxValue=255.0,
    adaptiveMethod=cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    thresholdType=cv2.THRESH_BINARY_INV,
    blockSize=19,
    C=9)
img_morphology5_blur = cv2.GaussianBlur(gray3, ksize=(5, 5), sigmaX=0)
img_morphology5_open= cv2.morphologyEx(img_morphology5_blur, cv2.MORPH_OPEN, kernel5)
img_morphology5_close = cv2.morphologyEx(img_morphology5_open, cv2.MORPH_CLOSE, kernel5)
img_morphology5_thresh_3 = cv2.adaptiveThreshold(
    img_morphology5_close,
    maxValue=255.0,
    adaptiveMethod=cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    thresholdType=cv2.THRESH_BINARY_INV,
    blockSize=19,
    C=9)
img_morphology5_blur = cv2.GaussianBlur(gray4, ksize=(5, 5), sigmaX=0)
img_morphology5_open= cv2.morphologyEx(img_morphology5_blur, cv2.MORPH_OPEN, kernel5)
img_morphology5_close = cv2.morphologyEx(img_morphology5_open, cv2.MORPH_CLOSE, kernel5)
img_morphology5_thresh_4 = cv2.adaptiveThreshold(
    img_morphology5_close,
    maxValue=255.0,
    adaptiveMethod=cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    thresholdType=cv2.THRESH_BINARY_INV,
    blockSize=19,
    C=9)

plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.title('test_img_5')
plt.imshow(img_morphology5_thresh_1, cmap='gray')
plt.subplot(2,2,2)
plt.title('test_img_6')
plt.imshow(img_morphology5_thresh_2, cmap='gray')
plt.subplot(2,2,3)
plt.title('test_img_7')
plt.imshow(img_morphology5_thresh_3, cmap='gray')
plt.subplot(2,2,4)
plt.title('test_img_8')
plt.imshow(img_morphology5_thresh_4, cmap='gray')

pytesseract.pytesseract.tesseract_cmd = 'C:\\tesseract\\tesseract\\tesseract'
config = ('-l kor --oem 3 --psm 7')
morphology_text_1 = pytesseract.image_to_string(img_morphology5_thresh_1, config=config)
print("morphology_1 text : " + morphology_text_1)
morphology_text_2 = pytesseract.image_to_string(img_morphology5_thresh_2, config=config)
print("morphology_2 text : " + morphology_text_2)
morphology_text_3 = pytesseract.image_to_string(img_morphology5_thresh_3, config=config)
print("morphology_3 text : " + morphology_text_3)
morphology_text_4 = pytesseract.image_to_string(img_morphology5_thresh_4, config=config)
print("morphology_4 text : " + morphology_text_4)
```
